```python
# Import Libraries
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.neighbors import KNeighborsRegressor
from sklearn.tree import DecisionTreeRegressor
from sklearn.svm import SVR
```

```python
# Read in csv files
lrc = pd.read_csv('LR_Column.csv')
mc = pd.read_csv('Mean_Column.csv')
medc = pd.read_csv('Median_Column.csv')
rc = pd.read_csv('Remove_Column.csv')
```

```python
# Models used for evaluation assigned to variable f as a list
f = [lrc,mc,medc,rc]

# Models used for evaluation assigned to variable m as a list
m = [LinearRegression(), KNeighborsRegressor(), SVR()
        ,DecisionTreeRegressor()]

# Title identifiers of the files
r = ['Linear Regression Column', 'Mean Column', 'Median Column', 'Remove Column']
```

```python
# Import MinMaxScaler and Train Test Split
from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split

# Create an instance of MinMaxScaler
scaler = MinMaxScaler()
```

```python
def files(df):
    # Identify X and Y variables
    X = df.drop('price',axis=1)
    y = df['price']
    
    # Train Test Split
    X_train, X_test, y_train, y_test=train_test_split(X, y, test_size=0.10, random_state=42)
    
    # Applying normalization 
    X_train = scaler.fit_transform(X_train)
    X_test = scaler.transform(X_test)
    
    for model in m:
        model.fit(X_train,y_train)
        y_pred = model.predict(X_test)
        print(f"Model used is {model}")
        print(f"Mean: {y_test.mean():.2f}")
        print(f"Mean Squared Error: {mean_squared_error(y_test,y_pred)}")
        print(f"Mean Absolute Error: {mean_absolute_error(y_test,y_pred)}")
        print(f"Root Mean Squared Error: {np.sqrt(mean_squared_error(y_test,y_pred))}")
        print(f"R2 Score: {(r2_score(y_test,y_pred)*100):.2f}")
        print("-------------------------------------------------------")
```

```python
for i in range(len(f)):
    print(f"\033[1m{r[i]}\033[0m")
    files(f[i])
    print('\n')
```
